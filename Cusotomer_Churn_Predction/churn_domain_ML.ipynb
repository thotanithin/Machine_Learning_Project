## Data Preparation and Model Building

```python
## Identifying input varaiable(X) and target varaiable(y)
X = churn[['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges']]
y = churn['Churn']
```

```python
y.head()
```

```python
X.head()
```

```python
## spliting data train and  test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25 ,random_state = 60)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
X_train.head()
```

```python
## separating  numerical and categorical columns
X_train_num = X_train.select_dtypes(include = ['int64','float64'])
X_train_cat = X_train.select_dtypes(include = "object")
```

```python
X_train_num.head()
```

```python
X_train_cat.head()
```

```python
X_train_cat = X_train_cat.drop("customerID",axis =1)
```

```python
X_train_cat.head()
```

## Data preparation of train data

## rescaling the numerical column

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_num_rescale   = pd.DataFrame(scaler.fit_transform(X_train_num),
        columns = X_train_num.columns,
        index  = X_train_num.index)
X_train_num_rescale.head()
```

```python
print("number of numerical Features :",scaler.n_features_in_)
print("Mean of numerical Features",scaler.mean_)
print("standard deviation of numerical features",np.sqrt(scaler.var_))
```

## Appling one hot encoding on categorical columns

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop = "first" ,sparse=False)
X_train_cat_ohe = pd.DataFrame(encoder.fit_transform(X_train_cat),
                               columns = encoder.get_feature_names_out(X_train_cat.columns),
                               index = X_train_cat.index)
X_train_cat_ohe.head()
```

## Concat the categorical columns and numerical columns of train data

```python
X_train_transformed = pd.concat([X_train_num_rescale,X_train_cat_ohe],axis =1)
X_train_transformed
```

## Data preparation of test data

```python
X_test.head()
```

```python
X_test.dtypes
```

```python
X_test_num = X_test.select_dtypes(include = ["int64",'float64'])
X_test_num.head()
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_cat.head()
```

```python
X_test_cat = X_test_cat.drop("customerID",axis =1)
```

```python

X_test_cat.head()
```

## rescaling the numerical column of test data


```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num),
                                   columns = X_test_num.columns,
                                   index = X_test_num.index)

X_test_num_rescaled.head()
```

## One hot encoder of test data

```python
X_test_cat_ohe = pd.DataFrame(encoder.transform(X_test_cat),
                              columns = encoder.get_feature_names_out(X_test_cat.columns),
                              index = X_test_cat.index)
X_test_cat_ohe.head()
```

## Concat the categorical columns and numerical columns of test data

```python
X_test_transformed = pd.concat([X_test_num_rescaled ,X_test_cat_ohe],axis =1)
X_test_transformed.head()
```

## Predection & evaulation of model

### Logistic Regression Classifier

```python
# Traning model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train_transformed,y_train)
```

```python
# predection model
y_test_pred = classifier.predict(X_test_transformed)
```

```python
# Evalution of model
from sklearn import metrics
metrics.accuracy_score(y_test,y_test_pred)
```

```python
sns.histplot(y_test, color='blue', alpha=0.5)
sns.histplot(y_test_pred, color='red', alpha=0.5)
```

##  DecisionTreeClassifier
